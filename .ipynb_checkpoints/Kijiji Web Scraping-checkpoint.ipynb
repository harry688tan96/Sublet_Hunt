# Kijiji Web Scraping On Room Sublet

```python
from urllib.request import urlopen
from urllib.parse import urlencode
from bs4 import BeautifulSoup
import re
import time
from geopy.geocoders import GoogleV3
from geopy.distance import vincenty
from geopy.exc import GeocoderTimedOut, GeocoderQuotaExceeded
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML
from multiprocessing import Pool
import sys

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
distance_radius = 3.0 # Limit the search for room sublets that are within 3.0 km
no_of_scraped_pages = 20 # Scrape a total of 20 pages in Kijiji
```

### Scape data from each specified Kijiji page

```python
def encodeURL(page):
    pageNum = ""
    if page != 1:
        pageNum = "/page-" + str(page)
   
    params = urlencode({
                        "address": "200 University Ave W, Waterloo, ON N2L 3G1", 
                        "ll": "43.4723,-80.5449"
                       })
    return "https://www.kijiji.ca/b-room-rental-roommate/kitchener-waterloo{0}/c36l1700212r{1}?{2}" \
            .format(pageNum, str(distance_radius), params)

    
url_to_ads = {}
def scrapeKijijiData(pageNum):
    print("Start scraping data from page {}".format(pageNum))
    
    url = encodeURL(pageNum)
    kijiji_page = urlopen(url)
    assert(kijiji_page.getcode() == 200)
    
    soup = BeautifulSoup(kijiji_page, "html.parser")    
    kijiji_page.close()
    
    tags_list = soup.findAll("div", {"class": "info"})
    
    for i, tag in enumerate(tags_list):
        price = tag.find("div", {"class": "price"}).text.strip()
        title_tag = tag.find("div", {"class": "title"})
        post_url = "https://www.kijiji.ca{}".format(title_tag.find("a")["href"])
        title = title_tag.text.strip()

        post_data = {
                        "price": price, 
                        "title": title, 
                        "post_url": post_url
                    }
                
        url_to_ads[post_url] = post_data
        
    print("Finished scraping page {}\n".format(pageNum))
```

**scrapePages** function scrapes Kijiji pages based on the number of pages that are needed to be scraped.

```python
def scrapePages(noOfPages):
    print("BEGIN SCRAPING\n-------------\n")
    
    for page in range(1, noOfPages + 1):
        scrapeKijijiData(page)
    
    print("\n---------------\nEND SCRAPING")
```

```python
scrapePages(no_of_scraped_pages)
```

### Extracting more details for each advertisement (ie. The date of the advertisement posted, the location of the house/apartment) by entering the hyperlink of each advertisement.

```python
def getMoreHousingDetails(url):
    post_data = url_to_ads[url]

    try:
        ads_page = urlopen(url)
        assert(ads_page.getcode() == 200)

        soup = BeautifulSoup(ads_page, "html.parser")
        ads_page.close()    # Important to close the page after the reading is done

        ads_metadata = soup.find("div", {"class":  re.compile(r"^itemMeta-*")})
        date_posted_data = ads_metadata.findAll("time")
        address_data = ads_metadata.findAll("span", {"class": re.compile(r"^address-*")})

        if len(date_posted_data):
            post_data["date posted (GMT)"] = date_posted_data[0]["title"]
        if len(address_data):
            post_data["address"] = address_data[0].text.strip()

        return post_data
        
    except:
        # Return an empty dictionary for URL that no longer exists (ie. the housing advertisement is no longer available)
        return {}
```

Opening a huge list of URLs sequentially is very slow. Instead, we should exploit the benefit of concurrency by ** MULTIPROCESSING ** . Using ** MULTIPROCESSING ** can achieve up to ** 10x ** faster.

```python
def openUrls(ads_links, multiprocessing=False, process_num=20, size=None):
    ads_posts = []
    ads_batch = ads_links[:size]
    
    if multiprocessing:
        with Pool(processes=process_num) as p:
            ads_posts = p.map(getMoreHousingDetails, ads_batch)
            p.terminate()
            p.join()
        
    else:
        for url in ads_batch:
            ads_posts.append(getMoreHousingDetails(url))
            
    return ads_posts
```

Experiment on **50** URLs **with** and **without** multiprocessing

```python
ads_links = list(url_to_ads.keys())
test_size = 50

tic = time.time()
ads_posts = openUrls(ads_links, size=test_size)
toc = time.time()
print("Total execution time on {} URLs WITHOUT multiprocessing: {}s".format(test_size, toc - tic))

tic = time.time()
ads_posts = openUrls(ads_links, multiprocessing=True, size=test_size)
toc = time.time()
print("Total execution time on {} URLs WITH multiprocessing:  {}s".format(test_size, toc - tic))
```

Open **ALL** URLs with multiprocessing

```python
tic = time.time()
ads_posts = openUrls(ads_links, multiprocessing=True)
toc = time.time()
print("Total execution time on ALL {} URLs WITH multiprocessing:  {}s".format(len(ads_links), toc - tic))
```

# The housing advertisements dataframe

First, drop rows that have missing data. Also, get rid of advertisements that are not offering a place to stay (ie. some advertisements are looking for places to stay). Then, reset the index for each row in place.

```python
kijiji_housing_df = pd.DataFrame(ads_posts)

# Drop rows that have missing data
kijiji_housing_df.dropna(axis=0, inplace=True)

# Get rid of ads that are not offering a place to stay
wanted_mask = kijiji_housing_df["title"].str.startswith("Wanted:")
kijiji_housing_df = kijiji_housing_df[~wanted_mask]

# Reset index in place
kijiji_housing_df.reset_index(drop=True, inplace=True)

kijiji_housing_df
```

### Include distance for each of the data
** Vincenty's formulae ** is used to calculate the ** distance ** (km) between the location of the house/apartment and the location of the University of Waterloo.

```python
def calculate_distance(address):
    geolocator = GoogleV3() # Google Maps API
    try:
        housing = geolocator.geocode(address)
        housing_coordinates = (housing.latitude, housing.longitude)
    
    except GeocoderTimedOut:
        return calculate_distance(address)
    
    except GeocoderQuotaExceeded:
        # Google Maps (standard API) only allows:
        #     1.) 2,500 free requests per day, calculated as the sum of client-side and server-side queries.
        #     2.) 50 requests per second, calculated as the sum of client-side and server-side queries.
        return calculate_distance(address)
    
    except:
        print ("Unexpected error:", sys.exc_info()[0])  # Hack to check the type of exception thrown by Geocoder
        return "Error in finding the coordinates for address: %s" % address

    UW_coordinates = (43.469757, -80.5409518)
    
    distance = vincenty(housing_coordinates, UW_coordinates).kilometers
    return round(distance, 3)
```

Again, we utilize the benefit of multiprocessing. 
First, we ** slice ** the dataframe into few chunks (ie. each chunk of size 10). Then, we apply the * calculate_distance * function on these chunks concurrently using multiple processors.
Again, there is a **10x** speed faster using multiprocessing.

```python
###### Apply function on DataFrame in parallel Using Multiprocessing #######
# https://stackoverflow.com/questions/26187759/parallelize-apply-after-pandas-groupby/29281494#29281494

process_num = 20

def applyParallel(dfGrouped, func):
    grouped_list = [group for _, group in dfGrouped]
    
    with Pool(processes=process_num) as p:
        ret_list = p.map(func, grouped_list)
        p.terminate()
        p.join()
    return pd.concat(ret_list)

def apply_row_df(df):
    df["distance (km)"] =  df['address'].apply(calculate_distance)
    return df

grouped = kijiji_housing_df.groupby(kijiji_housing_df.index // 10) # Partition into chunks of size 10

start_execute = time.time()
kijiji_housing_df = applyParallel(grouped, apply_row_df)
finish_execute = time.time()

print("Total execution time using multiprocessing:  {}s".format(finish_execute - start_execute))
```

# Graph Plotting

Convert the prices which are in **string** format to **float** format for graph plotting.

```python
pd.set_option('display.max_colwidth', 50) # Set the width of each columns back to default value, 50, to make it looks tidier

def extractPriceValue(price):
    # When there is no price listed for a housing post
    if (price.find("$") == -1):
        return 0
    
    # When there are both new price and old price listed in the price section, we only get the new price
    if (price.find("\n") != -1):
        price = price.split("\n")[0]
    
    return float(price[1:].replace(",", ""))

distance_and_price = kijiji_housing_df[["distance (km)", "price"]].copy()
distance_and_price.fillna("", inplace=True)
distance_and_price["price"] = distance_and_price["price"].apply(extractPriceValue)
```

```python
# Plot the graph (price vs distance (km))
distance_and_price.plot(x="distance (km)", y="price", kind="scatter")
plt.show()
```
